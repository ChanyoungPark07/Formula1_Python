```python
from matplotlib import pyplot as plt
import fastf1 as ff1
import fastf1.plotting
```

```python
fastf1.plotting.setup_mpl()
ff1.Cache.enable_cache('cache') 
```

```python
session = fastf1.get_session(2022, 'Austin', 'Q')
session.load()
```

```python
ham_lap = session.laps.pick_driver('HAM').pick_fastest()
ham_car_data = ham_lap.get_car_data()

ver_lap = session.laps.pick_driver('VER').pick_fastest()
ver_car_data = ver_lap.get_car_data()
```

```python
ham_time = ham_car_data['Time']
ham_speed = ham_car_data['Speed']

ver_time = ver_car_data['Time']
ver_speed = ver_car_data['Speed']
```

```python
fig, ax = plt.subplots()

ax.plot(ham_time, ham_speed, color='lime', label='HAM')
ax.plot(ver_time, ver_speed, color='blue', label='VER')
ax.set_xlabel('Time')
ax.set_ylabel('Speed [Km/h]')
ax.set_title('US Grand Prix Qualifying HAM vs. VER')
ax.legend()
plt.show()
```
